```python
import pandas as pd

# 데이터프레임 로드
# 예: dataset = pd.read_csv("your_dataset.csv")

# 고유한 레이블 값 확인
unique_labels = sorted(dataset['label'].unique())
print("Unique labels before remapping:", unique_labels)

# 레이블 값을 매핑
label_mapping = {label: idx for idx, label in enumerate(unique_labels)}
print("Label mapping:", label_mapping)

# 레이블 값 변환
dataset['label'] = dataset['label'].map(label_mapping)

# 변환된 레이블 확인
print("Unique labels after remapping:", dataset['label'].unique())
```

```python
# 데이터셋을 Hugging Face Dataset 객체로 변환
from datasets import Dataset
dataset = Dataset.from_pandas(dataset)
```

```python
# 데이터셋 분리 (train/test split)
train_test = dataset.train_test_split(test_size=0.2)
train_set = train_test['train']
test_set = train_test['test']
```

```python
# 모델 load
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')
```

```python
# bert 입력값으로 넣은거 1. 토크나이징  2. 그 토큰화된 거에 3ㅏ지 입베딩  (이건 사전학습되어서 이미 있고)
# 우리는 토크나이저만 주면됨

tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
```

## 전처리

```python
# label을 정수형으로 변환
train_set = train_set.map(lambda x: {'label': int(x['label'])})
test_set = test_set.map(lambda x: {'label': int(x['label'])})
```

```python
# 이거 항상 쓸거니까 암기하자 -> 교수님
# 전처리
# 256 넘어가면 text 잘라버린다는 옵션

# 토큰화 함수 정의
def preprocess(examples):
  return tokenizer(examples['preprocessed_text'],truncation=True, padding='max_length', max_length=256)

# 데이터셋에 전처리 적용
train_set = train_set.map(preprocess, batched=True, batch_size=16)
test_set = test_set.map(preprocess, batched=True, batch_size=16)
```

```python
train_set[0]
```

```python
# 텐서 형식으로 변환 (input_ids, attention_mask, label)
train_set.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])
test_set.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])
```

```python
train_set[0]
```

## 2. Training

```python
from transformers import TrainingArguments

# TrainingArguments 설정
training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/2024_3학년_2학기/데이터마이닝과파이썬프로그래밍/results",           # 출력 디렉토리
    evaluation_strategy="epoch",      # 매 에포크마다 평가
    save_strategy="epoch",            # 매 에포크마다 모델 저장
    learning_rate=2e-5,               # 학습률
    per_device_train_batch_size=16,   # 배치 크기
    per_device_eval_batch_size=16,    # 평가 시 배치 크기
    num_train_epochs=10,               # 에포크 수
    weight_decay=0.01,                # 가중치 감쇠
    logging_dir="./logs",             # 로그 저장 경로
    logging_steps=10,                 # 로그 출력 주기
    save_total_limit=2,               # 저장할 체크포인트 개수 제한
    report_to="none"                  # 로그를 어디로 보낼지 (기본값: wandb)
)
```

```python
# 평가지표

from sklearn.metrics import f1_score, accuracy_score

# 평가 메트릭 함수 정의
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)  # 가장 높은 확률을 가진 클래스 예측
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds, average="weighted")
    }
```

```python
from transformers import BertForSequenceClassification

# BERT 모델 초기화
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=13)
```

```python
from transformers import Trainer

# Trainer 초기화
trainer = Trainer(
    model=model,                         # 모델
    args=training_args,                  # 학습 설정
    train_dataset=train_set,             # 학습 데이터셋
    eval_dataset=test_set,               # 평가 데이터셋
    compute_metrics=compute_metrics      # 평가 메트릭
)
```

```python
# 학습 시작
trainer.train()
```

```python
# 평가
eval_results = trainer.evaluate()

# 결과 출력
print(eval_results)
```
